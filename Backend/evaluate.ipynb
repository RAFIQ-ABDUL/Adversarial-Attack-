```python
import tensorflow as tf
import numpy as np
from fgsm import FGSM_Attack
```

```python
model = tf.keras.models.load_model("mnist_model.h5")
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_test = x_test / 255.0
x_test = x_test[..., tf.newaxis]
```

```python
attack = FGSM_Attack(model)
```

```python
correct = 0
epsilon = 0.1

for i in range(1000):
    image = x_test[i:i+1]
    label = y_test[i:i+1]

    adv = attack.generate(image, label, epsilon)
    pred = tf.argmax(model(adv), axis=1)

    if pred == label:
        correct += 1

print("Accuracy under attack:", correct/1000)
```
